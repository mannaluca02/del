```python
from datetime import datetime
import numpy as np
import torch
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
from torchsummary import summary
import matplotlib.pyplot as plt
```

#### Loading the Data

```python
training_data = datasets.FashionMNIST(
    root="../data",
    train=True,
    download=True,
    transform=ToTensor()
)
test_data = datasets.FashionMNIST(
    root="../data",
    train=False,
    download=True,
    transform=ToTensor()
)
print(training_data)
print(test_data)
```

```python
img, label = training_data[0]
plt.imshow(img.squeeze(), cmap="gray")
plt.title(f"Label: {label}")
plt.show()
```

```python
import math
def plot_training(nepochs, cost_hist, cost_hist_test, acc_hist, acc_hist_test, cost_range=None, acc_range=None):
    plt.figure(1)
    plt.plot(torch.arange(nepochs), cost_hist, "-", label='train')
    plt.plot(torch.arange(nepochs), cost_hist_test, "r-", label='test')
    if cost_range:
        mincost, maxcost = cost_range
    else: 
        mincost, maxcost = 0.0, max(acc_hist)
    plt.xlabel("Epoch")
    plt.xlim(0,nepochs)
    plt.ylim(mincost, maxcost)
    plt.title("Cross-Entropy Cost")
    plt.legend()
    plt.figure(2)
    plt.plot(torch.arange(nepochs), acc_hist,"-", label='train')
    plt.plot(torch.arange(nepochs), acc_hist_test,"r-", label='test')
    if acc_range:
        minacc, maxacc = acc_range 
    else:
        min_acc, maxacc = min(min(acc_hist),0.8), 1.0
    plt.xlabel("Epoch")
    plt.xlim(0,nepochs)
    plt.ylim(minacc,maxacc)
    plt.legend()
    plt.title("Accuracy")
```

#### CNN

```python
class CNN(torch.nn.Module):
    
    def __init__(self, nhidden1=256, nhidden2=64, noutput=10):
        super(CNN, self).__init__()
        self.layers = torch.nn.Sequential(
            torch.nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1, bias=True),
            torch.nn.ReLU(),
            torch.nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, padding=1, bias=True),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
            torch.nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1, bias=True),
            torch.nn.ReLU(),
            torch.nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1, bias=True),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
            torch.nn.Flatten(),
            torch.nn.Linear(7*7*64, nhidden1),
            torch.nn.ReLU(),
            torch.nn.Linear(nhidden1, nhidden2),
            torch.nn.ReLU(),
            torch.nn.Linear(nhidden2, noutput)
        ) 
    
    def forward(self, x):
        z = self.layers(x)
        return z
```

#### GPU

```python
device = None
if torch.cuda.is_available():
    device = torch.device("cuda")
elif torch.backends.mps.is_available():
    device = torch.device("mps")
else: 
    device = torch.device("cpu")

print(device)
```

```python
def train_val(model, cost, epochs, train_loader, val_loader, optimizer, device):
    model = model.to(device)
    cost = cost.to(device)
    
    cost_hist = []
    cost_hist_test = []
    acc_hist = []
    acc_hist_test = []    
    train_samples = len(train_loader.dataset)
    val_samples = len(val_loader.dataset)
    train_batches = len(train_loader)
    val_batches = len(val_loader)
    for epoch in range(nepochs):
        cost, acc = 0.0, 0.0
        for batch, (X, Y) in enumerate(train_loader):
            X, Y = X.to(device), Y.to(device)
            pred = model(X)
            loss = cost_ce(pred, Y)
            cost += loss.item()
            acc += (pred.argmax(dim=1) == Y).type(torch.float).sum().item()
    
            # gradient, parameter update
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        cost /= train_batches
        acc /= train_samples
    
        cost_val, acc_val = 0.0, 0.0
        with torch.no_grad():
            for batch, (X, Y) in enumerate(val_loader): 
                X, Y = X.to(device), Y.to(device)
                pred = model(X)
                cost_val += cost_ce(pred, Y).item()
                acc_val += (pred.argmax(dim=1) == Y).type(torch.float).sum().item()
            cost_val /= val_batches
            acc_val /= val_samples
        
        print("Epoch %i: %f, %f, %f, %f"%(epoch, cost, cost_val, acc, acc_val))
        cost_hist.append(cost)
        cost_hist_test.append(cost_val)
        acc_hist.append(acc)
        acc_hist_test.append(acc_val)

    return cost_hist, cost_hist_test, acc_hist, acc_hist_test
```

```python
nbatch = 64
nepochs = 20
learning_rate = 0.02

model = CNN()
cost_ce = torch.nn.CrossEntropyLoss()

train_loader = DataLoader(training_data, batch_size=nbatch, shuffle=True)
test_loader = DataLoader(test_data, batch_size=nbatch, shuffle=False)

sgd = torch.optim.SGD(model.parameters(), lr=learning_rate)

cost_hist1, cost_hist_test1, acc_hist1, acc_hist_test1 = train_val(model, cost_ce, nepochs, train_loader, test_loader, sgd, device)
```

```python
plot_training(nepochs, cost_hist1, cost_hist_test1, acc_hist1, acc_hist_test1, cost_range=(0.0, 5.0), acc_range=(0.0,1.0))
```

### Standard Normal Weights for Conv Layers

```python
nbatch = 64
nepochs = 20
learning_rate = 0.02

def init_weights(m):
    if isinstance(m, torch.nn.Conv2d):
        torch.nn.init.normal_(m.weight, 0.0, 1.0)

model1 = CNN()
model1.apply(init_weights)

cost_ce = torch.nn.CrossEntropyLoss()

train_loader = DataLoader(training_data, batch_size=nbatch, shuffle=True)
test_loader = DataLoader(test_data, batch_size=nbatch, shuffle=False)

sgd = torch.optim.SGD(model1.parameters(), lr=learning_rate)

cost_hist2, cost_hist_test2, acc_hist2, acc_hist_test2 = train_val(model1, cost_ce, nepochs, train_loader, test_loader, sgd, device)
```

```python
plot_training(nepochs, cost_hist2, cost_hist_test2, acc_hist2, acc_hist_test2, cost_range=(0.0, 3.0), acc_range=(0.0,1.0))
```
