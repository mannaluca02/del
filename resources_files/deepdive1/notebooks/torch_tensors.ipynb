```python
import torch
```

#### Initialization

```python
# from lists
x = torch.Tensor([5, 3])
x, x.shape, x.dtype, x.device
```

```python
# with zeros or ones
x = torch.zeros([3, 2])
x, x.shape
```

```python
y = torch.ones([2, 3])
y, y.shape
```

```python
# random uniform 
x = torch.rand([2, 5])
x, x.shape
```

```python
# random normal
y = torch.randn([2, 5])
y, y.shape
```

#### Operations

```python
x = torch.Tensor([5, 3, 1])
y = torch.Tensor([2, 4, 8])
```

```python
# element-wise operations
print('Addition: ', x + y)
print('Subtraction: ', x - y)
print('Multiplication: ', x * y)
print('Division: ', x / y)
print((x-y).abs())
```

```python
# math reduction
x = torch.randn([2, 5])
print(x)
print(x.sum())
print(x.abs().sum())
print(x.sum(dim=1))
print(x.max().item()) # for 0d tensors
print(x.min(dim=1))
```

```python
# linear algebra operations
x = torch.tensor([ 10., 20., 30.])
m = torch.Tensor([[0,0,1],[0,2,0],[5,0,0]])
print(m.mv(x))
print(m @ x)
```

```python
print(m @ m.T)
```

```python
torch.inverse(m @ m.T)
```

#### torch <-> numpy

```python
import numpy as np
```

```python
xnp = np.array([[1,2],[3,4]])
x = torch.Tensor(xnp)
x
```

```python
x[0,1]=5
xnp[1,0]=6
x, xnp
```

```python
xx = torch.from_numpy(xnp)
xx[0,1]=5
xnp[1,0]=7
xx, xnp
```

```python
xxnp=xx.numpy()
xx[0,1]=8
xxnp[1,0]=9
xx, xxnp
```

#### Reshape

```python
# Reshaping tensors
x = torch.rand([3, 4])
y = x.view([2, 6])
z = x.view(-1)
x, y, z
```

```python
z[0]=10
x, y, z
```

```python
xx = torch.rand([3, 4])
zz = xx.reshape(2,-1)
zz[0,1]=10
xx, zz
```
